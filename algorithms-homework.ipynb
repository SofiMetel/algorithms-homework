```python
def isPalindrome(str):
  
    # Run loop from 0 to len/2
    for i in range(0, int(len(str)/2)):
        if str[i] != str[len(str)-i-1]:
            return False
    return True

isPalindrome(word)
```

```python
#1. Write an algorithm that returns the largest element in an array. Assume the array is unsorted.

arr = [0, 42, 2, 8, 10, 7]

def max(arr):
    maxnum=0
    for i in arr:
        if i> maxnum:
            maxnum = i
    return maxnum
            
max(arr)
```

2. Write an algorithm that returns a new array which is the reverse of the input.

```python
arr = [1, 42, 2, 8, 10, 7]

def reverse(arr):
    n = len(arr)
    revarr = []
    for i in range(len(arr)):
        revarr.append(arr[n-i-1])
    return revarr

reverse(arr)
```

```python
#4. Write an algorithm that returns the elements on odd positions in an array.

arr1 = [1, 42, 2, 8, 10, 7, 9, 99]

def oddposition(arr):
    i=0
    newarr = []
    while i<len(arr):
        if i%2!=1:
            i=i+1
        else:
            newarr.append(arr[i])
            i=i+1

    return newarr


print(oddposition(arr1))
```
